### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

ANSWER ( part 1 ): 

In Matplotlib, plotting can be done in two ways: 
A.The Pyplot (State-based) approach in Matplotlib is a simpler way to create plots. It works like how plotting is done in MATLAB. We donâ€™t need to manually create figures or axes; instead, we just call functions like plt.plot() or plt.scatter(). Matplotlib automatically takes care of creating the figure and axes for us in the background. This approach is quick and easy, making it great for creating simple, one-time plots without needing to manage the plot's details.





```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4], [15, 25, 35, 40])
plt.title("Pyplot Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Answer(Part 2):

B. The Object-oriented approach in Matplotlib is another  way to create plots. 
In this method, we first create a figure and axes, which act like containers for the plot. 
Then  it plots the data directly onto the axes and have full control over customizing the plot,
such as setting axis labels, limits, and adding legends. 
This approach is more flexible and works well for creating complex plots, especially when we need multiple plots in one figure (like subplots).

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [15, 25, 35, 40])
ax.set_title("Object-Oriented Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")

plt.show()
```

ANSWER ( Final ): 
 
 The recommended approach is the Object-oriented approach. It's better for more complex plots because it gives us more control and flexibility, especially when we have multiple plots or need to customize them in detail. This approach is often used in professional code because it allows us to manage all aspects of the plots. While the Pyplot approach is faster and easier for simple plots, the Object-oriented approach is the best choice for creating advanced visualizations with precise control over every element.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

ANSWER :
##
1.The figure is the very basic foundation for plotting data in matplotlib.It is essentially the entire plotting area that holds everythingâ€”axes, titles, lines,dots,bars,lables other visual elements. A figure can contain one or more axes.

2.The axes is the area where plot appears in. It is the region within the figure where the data is plotted. It defines the area for displaying data points, as well as labels, ticks, and gridlines. A figure can have multiple axes (for example, in subplots), and each axes corresponds to a single plot.

3.Axis is one of the coordinate axes of the plot. It  refers to the individual x-axis and y-axis (and possibly z-axis in 3D plots) within the axes. These axes define the coordinate system, including the range of values displayed and the tick marks. Each axes object contains an x-axis and a y-axis, each having their own properties like labels, tick positions, and limits.

4.All objects that interact with on a Matplotlib plot are called Artist .An artist in Matplotlib is any element that can be drawn or rendered on the plot. This includes graphical elements like lines, markers, text, patches (such as rectangles or circles), images, and legends. Artists are the components that visually represent the data or provide information on the plot.

### When plotting in Matplotlib, what is the expected input data type?

ANSWER
##
In Matplotlib, the expected input data type for plotting is typically a numeric array-like structure. This means we can provide:
1.Lists: Standard Python lists containing numerical data.

2.NumPy arrays: Arrays from the NumPy library, which are efficient and commonly used for numerical data

3.Pandas Series or DataFrames: When we are using Pandas, then we can directly pass Series (for one-dimensional data) or DataFrames (for two-dimensional data) to Matplotlib.

Matplotlib can handle all of these data types and will treat them as numbers to create plots. The data needs to be numeric (like integers or floats) because Matplotlib is built to visualize numerical information.









    


### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(-4, 4, 100)  # Generate 100 points from -4 to 4
y = x**2                     # Compute y = x^2

# Object-Oriented Plot
fig, ax = plt.subplots()     # Create figure and axes
ax.plot(x, y, label="y = x^2", color="blue")  # Plot the data
ax.set_title("Title - Plot of y = x^2")              # Set title
ax.set_xlabel("x")                           # Set x-axis label
ax.set_ylabel("y")                           # Set y-axis label
ax.legend()                                  # Add legend
plt.show()                                   # Display the plot
```

```python
# Pyplot Plot
plt.plot(x, y, label="y = x^2", color="blue")  # Plot the data
plt.title("Title - Plot of y = x^2")                 # Set title
plt.xlabel("x")                              # Set x-axis label
plt.ylabel("y")                              # Set y-axis label
plt.legend()                                 # Add legend
plt.show()                                   # Display the plot
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
# Create the figure and subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Scatter plot
ax1.scatter(random_data_x, random_data_y, alpha=0.5, label="Random Data", color='blue')
#ax1.plot(x, y, color='red', label="y = x^2")
ax1.set_title("Scatter Plot")
ax1.set_xlabel("X values")
ax1.set_ylabel("Y values")
ax1.legend()

# Bar plot
ax2.bar(names, values, color='green')
ax2.set_title("Bar Plot of Fruits")
ax2.set_xlabel("Fruit Names")
ax2.set_ylabel("Quantity")

# Adjust layout and display
plt.tight_layout()
plt.show()
```
